## 1. Variables and Data Types

### Variables
Variables are used to store data that can be used and manipulated throughout your program. You do not need to declare a variable before using it in Python.

### Data Types
Python has several built-in data types, including:
- **Integers**: Whole numbers, e.g., `10`
- **Floats**: Decimal numbers, e.g., `3.14`
- **Strings**: Sequence of characters, e.g., `"Hello"`
- **Booleans**: True or False values, e.g., `True`

# Operators in Python

## 1. Arithmetic Operators
Arithmetic operators are used to perform mathematical operations.

- `+` : Addition
- `-` : Subtraction
- `*` : Multiplication
- `/` : Division
- `%` : Modulus
- `**` : Exponentiation
- `//` : Floor division

## 2.Comparison Operators
Comparison operators are used to compare two values.

- `==` : Equal to
- `!=` : Not equal to
- `>` : Greater than
- `<` : Less than
- `>=` : Greater than or equal to
- `<=`  : Less than or equal to

## 3.Logical Operators
Logical operators are used to combine conditional statements.

- `and` : Returns True if both statements are true
- `or` : Returns True if one of the statements is true
- `not` : Reverses the result, returns False if the result is true

## Assignment Operators
Assignment operators are used to assign values to variables.

- `=`: Assign
- `+=` : Add and assign
- `-=` : Subtract and assign
- `*=` : Multiply and assign
- `/=` : Divide and assign
- `%=` : Modulus and assign
- `**=` : Exponentiation and assign
- `//=` : Floor division and assign

## Membership Operators
Membership operators are used to test if a sequence is presented in an object.

  -`in` : Returns True if a sequence with the specified value is present in the object<br/>
  -`not in` : Returns True if a sequence with the specified value is not present in the object

## Identity Operators
Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.

-`is` : Returns True if both variables are the same object<br/>
-`is not` : Returns True if both variables are not the same object
