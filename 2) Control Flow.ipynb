## Goal
The goal of this chapter is to understand how to control the flow of execution in a Python program.

 - conditional statements
 - loops


*By the end of this chapter, you should be able to use these control flow tools to create dynamic and flexible programs.*


## 1. Conditional Statements

### If Statement
The `if` statement allows you to execute a block of code only if a condition is true.

```python
age = 15

if age >= 18:
    print("You Can vote")
```

### If-Else Statement
The `if-else` statement allows you to execute one block of code if the condition is true, and another block if it is false.

```python
marks = 35

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

### Elif Statement
The `elif` statement stands for "else if" and allows you to check multiple conditions.

```python
traffic_signal = "Yellow"

if traffic_signal == "Green":
    print("Go")
elif traffic_signal =="Yellow":
    print("Slow Down")
else:
    print("Stop")
    
```

## 2. Loops

### While Loop
The `while` loop allows you to execute a block of code as long as a condition is true.

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

### For Loop
The `for` loop allows you to iterate over a sequence (such as a list, tuple, dictionary, set, or string).

```python
for i in range(5):
    print(i)
```

### Nested Loops
You can use nested loops to perform complex iterations.

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```

## 3. Loop Control Statements

### Break Statement
The `break` statement is used to exit a loop prematurely.

```python
for i in range(10):
    if i == 5:
        break
    print(i) 
        
```

### Continue Statement
The `continue` statement is used to skip the current iteration and continue with the next iteration of the loop.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)    
```

### Pass Statement
The `pass` statement is a null statement that is used as a placeholder.

```python
if True:
    pass
```
