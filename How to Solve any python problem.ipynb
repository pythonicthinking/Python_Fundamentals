# Problem Statement

Write a Python program to find the largest number in a list.

Example:

Input:

```python
[10, 5, 30, 8]
```

Output:

```python
30
```

---


# Step 1: Understand The Problem

Before writing any code, understand what the problem is asking.

### Questions to Ask

- What is the goal?
- What should the program return?
- Are there any special cases?

### In Our Example

We need to find the largest value from a list of numbers.

We do NOT need to:

- Sort the list
- Find the average
- Find the smallest number

Only find the largest number.

---

# Step 2: Identify Input and Output

### Input

```python
[10, 5, 30, 8]
```

### Output

```python
30
```

### Input-Output Table

| Input | Output |
|---------|---------|
| [10,5,30,8] | 30 |
| [100,50,400,200] | 400 |
| [-10,-20,-5] | -5 |

---


# Step 3: Break The Problem Into Smaller Steps

Let's write them.

1. Take first value.
2. Compare remaining values.
3. Update largest value.
4. Display answer.

---

# Step 4: Write The Algorithm

## Algorithm

1. Store the first element as largest.
2. Traverse through the list.
3. Compare each number with largest.
4. If larger, update largest.
5. Print largest.

### Pseudocode

```text
SET largest = first element

FOR each number in list

    IF number > largest

        largest = number

PRINT largest
```

---

# Step 5: Convert Algorithm Into Python Code

````python
numbers = [100,50,400,200]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num
        print(largest)
````
